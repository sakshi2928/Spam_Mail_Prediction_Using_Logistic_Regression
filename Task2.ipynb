**Importing Libraries**


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

**Loading dataset**

```python
df = pd.read_csv('/content/SampleSuperstore - SampleSuperstore.csv')
print(df)
```

**Visualizing the distribution of profits**

```python
plt.figure(figsize=(10, 6))
plt.hist(data=df, x='Profit')
plt.title('Profit Distribution')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

**We visualized the distribution of profits using a histogram with a kernel density estimate (KDE).**

```python
# Correlation between different variables (e.g., Profit and other factors)
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

**We visualized the correlation matrix to understand how different variables, such as profit and other factors, are related to each other.**

```python
# Scatter plot to visualize profit against another variable (e.g., sales)
plt.figure(figsize=(10, 6))
plt.scatter(df['Sales'], df['Profit'])
plt.title('Profit vs. Sales')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

**we cann conclude that there are many areas where we need to work for making profits.**
