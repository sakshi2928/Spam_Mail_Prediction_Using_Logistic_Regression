# ***UNIVARIATE ANALYSIS :***

Analyzing and comprehending one variable at a time while concentrating on its distribution, statistics, and visualization without taking into account relationships with other variables is known as univariate analysis


```python
edu=df['Education'].value_counts().reset_index()
edu
```

```python
sns.set_style("darkgrid")
colors = sns.color_palette("pastel")
plt.figure(figsize=(10, 6))
sns.barplot(x='index', y='Education', data=edu, palette=colors)
plt.xlabel('Education Level', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Education Level Distribution', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.tight_layout()
plt.show()
```

The barplot shows that most of the employee's are the one's with bachelors degree.

```python
city=df['City'].value_counts().reset_index()
city
```

```python

plt.figure(figsize=(12, 8))
ax = sns.barplot(x='index', y='City', data=city, palette='viridis')
plt.xlabel('City', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('City Distribution', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')
plt.tight_layout()
plt.show()
```

The above plotting shows that the most of the employee's are from bangalore city.

# ***BIVARIATE ANALYSIS***

The statistical examination of two variables at once to find correlations, differences, or links between them is known as bivariate analysis

```python
cross_tab = pd.crosstab(df['Education'], df['PaymentTier'])
custom_colors = ["#D891EE", "#C649EB", "#9A39B8"]
plt.figure(figsize=(10, 6))
cross_tab.plot(kind='bar', stacked=True, color=custom_colors)
plt.xlabel('Education')
plt.ylabel('Count')
plt.title('Distribution of PaymentTier by Education')
plt.xticks(rotation=45)
plt.legend(title='PaymentTier', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

The above stacked bar plot shows that employee's with bachelor degree are highly paid as compared to others.

```python

plt.scatter(df['Age'], df['ExperienceInCurrentDomain'])
plt.xlabel('Age')
plt.ylabel('Experience in Current Domain')
plt.title('Scatter Plot of Age vs. Experience in Current Domain')
plt.show()
```

The scatter plot shows that  2 to 4 employee's of age group 35 and 40 have more experience in their current domain than others.

```python

cross_tab_city_leave = pd.crosstab(df['City'], df['LeaveOrNot'])
plt.figure(figsize=(10, 6))
cross_tab_city_leave.plot(kind='bar', stacked=True, colormap='coolwarm')
plt.xlabel('City')
plt.ylabel('Count')
plt.title('Distribution of LeaveOrNot by City')
plt.xticks(rotation=45)
plt.legend(title='LeaveOrNot', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

The above stacked bar plotting shows that the employee's from bangalore city take less leave as compared to other two cities.

```python

numerical_columns = ['JoiningYear', 'Age', 'ExperienceInCurrentDomain']
numerical_data = df[numerical_columns]
correlation_matrix = numerical_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

**HEATMAP:**

The heatmap uses color intensity to represent the strength and direction of
correlations.

A positive correlation (close to 1) suggests that as one variable increases, the other tends to increase as well.

A negative correlation (close to -1) suggests that as one variable increases, the other tends to decrease.

A strong positive correlation between 'Age' and 'ExperienceInCurrentDomain' suggest  that as age increases, experience in the current domain tends to increase as well.

# ***MULTIVARIATE ANALYSIS***

Multivariate analysis involves the simultaneous analysis and interpretation of relationships among multiple variables within a dataset.

```python
sns.pairplot(df[['JoiningYear', 'Age', 'ExperienceInCurrentDomain']])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

The dataset can be explored visually with the pair plot, which enables a basic understanding of the relationships and distributions among various numerical variables.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Age', data=df)
plt.title('Box Plot of Age by Gender')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='ExperienceInCurrentDomain', data=df)
plt.title('Box Plot of ExperienceInCurrentDomain by Gender')
plt.show()
```

The first boxplot shows us that the proportion of females are more than males in the age group 26 to 32.

The second boxplot shows us that both male and female have equal proportion of experience in their current domain.

```python


# Grouped bar plot considering 'Education', 'PaymentTier', and 'Gender'
plt.figure(figsize=(12, 6))
sns.countplot(x='Education', hue='PaymentTier', data=df, palette='Set2', dodge=True)
plt.title('Distribution of PaymentTier by Education and Gender')
plt.xlabel('Education')
plt.ylabel('Count')
plt.legend(title='PaymentTier', loc='upper right', labels=['PaymentTier 1', 'PaymentTier 2', 'PaymentTier 3'])
plt.show()
```

We are considering 'Education' on the x-axis, 'PaymentTier' as hue (dividing the bars), and 'Gender' as an additional factor influencing the color within each category.

The grouped bars now represent the distribution of 'PaymentTier' for each 'Education' level, considering 'Gender' as an additional factor for differentiation.

This provides a multivariate perspective by incorporating multiple variables ('Education', 'PaymentTier', and 'Gender') into a single visualization.

The different shades or colors within each bar group indicate the presence of 'Gender' as a distinct category influencing the representation.
